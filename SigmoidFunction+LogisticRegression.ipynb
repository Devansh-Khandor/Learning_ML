# Sigmoid Function and Logistic Regression

```python
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt
```

## Sigmoid Function

```python
def sigmoid(z):
    g_z = 1/(1+np.exp(-z))
    return g_z
```

## Logistic Regression

```python
def logistic(x, w, b):
    f_wb = w*x+b
    f_wb = sigmoid(f_wb)
    return f_wb
```

## Applying above concepts

```python
x_train = np.array([0., 1, 2, 3, 4, 5])
y_train = np.array([0,  0, 0, 1, 1, 1])

w_in = 4
b_in = -10
```

```python
y_hat = logistic(x_train, w_in, b_in)
print(y_hat)
```

# Plotting

```python
fig, ax = plt.subplots(figsize=(12, 6))
plt.subplots_adjust(bottom=0.28)
# Smooth curve
x_curve = np.linspace(0, 10, 400)
y_curve = logistic(x_curve, w_in, b_in)

sigmoid_line, = ax.plot(
    x_curve, y_curve,
    linewidth=4,
    label="y = sigmoid(z)"
)

# Benign (0)
ax.scatter(
    x_train[y_train == 0],
    y_train[y_train == 0],
    facecolors='none',
    edgecolors='blue',
    s=140,
    linewidths=2,
    label="benign"
)

# Malignant (1)
ax.scatter(
    x_train[y_train == 1],
    y_train[y_train == 1],
    marker='x',
    color='red',
    s=140,
    linewidths=3,
    label="malignant"
)
```
